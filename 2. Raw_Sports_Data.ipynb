## Raw Data Frames for Each Sports Team

We wanted all this raw data in one place to compare the natural overlaps within the data to more easily make clean data frames for each sports team.

### 2018 Eagles Raw Data

```python
import pandas as pd
```

```python
eagles_tables = pd.read_html("https://www.statmuse.com/nfl/team/philadelphia-eagles-49/schedule/2018")
```

```python
eagles_df = eagles_tables[2]
```

```python
eagles_df.to_csv("../data/eagles.csv", index=False)
```

```python
eagles_df.head(10)
```

This Eagles data had the least amount of information for us to work with compared to the multiple columns of Phillies and 76ers. Consequently, the clean sports data frames (under Clean_Sports_Data.ipynb) reflect only the most important data that are reachable across all the teams. 

### 2018 Phillies Raw Data

```python
phillies_tables = pd.read_html("https://www.baseball-reference.com/teams/PHI/2018-schedule-scores.shtml")
```

```python
phillies_df = phillies_tables[0]
```

```python
phillies_df.to_csv("../data/phillies.csv", index=False)
```
